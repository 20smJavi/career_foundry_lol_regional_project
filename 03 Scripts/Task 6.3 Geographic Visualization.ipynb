## 02. Creating common column

Creating regions list

```python
BR = ['Brazil']
```

```python
EUN = ['Albania','Belarus','Bosnia and Herzegovina','Bulgaria','Croatia','Czechia','Czech Republic','Estonia','Finland','Greece','Hungary','Kosovo','Latvia','Lithuania','Moldova','Norway','Poland','Republic of Macedonia','Romania','Serbia','Slovakia','Sweden','Ukraine']
```

```python
EUW = ['Falkland Islands','Greenland','French Southern and Antarctic Lands','France','Austria','Germany','Portugal','Spain','Ireland','New Caledonia','Italy','Denmark','United Kingdom']
```

```python
KR = ['South Korea']
```

```python
LAN = ['Mexico','Guatemala','Honduras','Nicaragua','Costa Rica','Panama','Venezuela','Colombia','Ecuador','Peru','Cuba','Dominican Republic','Puerto Rico']
```

```python
LAS = ['Bolivia','Paraguay','Chile','Uruguay','Argentina']
```

```python
NA = ['United States of America', 'Canada']
```

```python
OC = ['Australia', 'New Zealand']
```

```python
RU = ['Russia']
```

```python
TR = ['Turkey']
```

Creating a new list to fill a server column for our world data

```python
# Creating server list
server = []

for value in world["NAME_EN"]:
  if value in BR:
    server.append('Brazil')
  elif value in EUN:
    server.append('Europe Nordic & East')
  elif value in EUW:
    server.append('Europe West')
  elif value in KR:
    server.append('Korea')
  elif value in LAN:
    server.append('Latin America North')
  elif value in LAS:
    server.append('Latin America South')
  elif value in NA:
    server.append('North America')
  elif value in OC:
    server.append('Oceania')
  elif value in RU:
    server.append('Russia')
  elif value in TR:
    server.append('Turkey')
  else:
    server.append('NaN')
```

```python
# Adding a new column and filling it with our new list
world['server'] = server
```

```python
# Doing a subset of world data
world_3 = world[['scalerank','SOVEREIGNT','SOV_A3','server','TYPE','NAME_EN','geometry']]
```

```python
world_3['server'].value_counts()
```

## 03. Data Merging

```python
df_merge = world_3.merge(df, on = 'server', how = 'outer', indicator = True)
```

```python
# Chekcing our shape
df_merge.shape
```

## 04. Data Wrangling

```python
# Checking for missing values
df_merge.isnull().sum()
```

```python
# Dropping any rows with NaN values, which would be any rows with 'server' NaN that we set up with our server column
df_merge.dropna(inplace = True)
```

```python
# Checking our merge value to find a full match indicated by having only 'both'
df_merge['_merge'].value_counts()
```

```python
# Dropping our merge columns
df_merge.drop(columns = '_merge', inplace = True)
```

```python
# Creating a data frame with our dups
df_merge_dups = df_merge.duplicated()
```

```python
# Checking for dupplicates
df_merge_dups.value_counts()
```

## 05. Creating a chropleth graph

```python
# Checking mean of 'game_duration' across the data set
df_merge['game_duration'].mean()
```

```python
# Creating a subset of games that are above the average (1613)
df_merge_durr = df_merge[df_merge['game_duration'] >= 1613]
```

```python
gd = geoplot.choropleth(
    df_merge_durr, hue = df_merge_durr['game_duration'], scheme = None,
    cmap='Blues', figsize=(15, 15)
)
```

The results of this graph, while difficult to interpret, indicates that servers 'Europe Nordic & East' and 'Europe West' have the longer games times, followed would be "North America, then 'Oceania'. This was one of the questions proposed initally and is answered via this graph.
