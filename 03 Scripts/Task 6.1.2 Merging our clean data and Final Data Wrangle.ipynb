#### Duplicating Column Data Types

Changing 2 columns, 'map_side' and 'outcome', while it might be easy to distinguish who got the first objective through the use of 0/1 booleans, analyzing the data and seeing "Team 0 Win 1" isn't easy to tell that "Red Side Won the match", so we are changing these data types and values to match what they would mean in a real world setting.

```python
merge_6 = merge_5.astype({'map_side' : 'object', 'outcome' : 'object'})
```

```python
merge_6.dtypes
```

Now to change the values within those columns

```python
merge_6['map_side'].replace({0:'Red Side'}, inplace = True)
```

```python
merge_6['map_side'].replace({1:'Blue Side'}, inplace = True)
```

```python
merge_6['outcome'].replace({0:'Defeat'}, inplace = True)
```

```python
merge_6['outcome'].replace({1:'Victory'}, inplace = True)
```
